# Systemic Tau v3.0 - Full Workflow

This notebook demonstrates the complete workflow in v3.0, transitioning from systemic tau calculation to topological layer extraction, and finally spatial hotspots.

```python
import numpy as np
import pandas as pd
from systemictau import ChaosGenerator, systemic_tau
from systemictau.layers import extract_joint_episodes, compute_antisynchronization, hyper_persistence, rolling_rqa, critical_mass_metric
import matplotlib.pyplot as plt

# 1. Generate Complex Data
X = ChaosGenerator.logistic_map_coupled(T=1000, N=10, coupling=0.15)
plt.plot(X[:100, 0], label='Node 1')
plt.plot(X[:100, 1], label='Node 2')
plt.title('Sample Complex Time Series')
plt.legend()
plt.show()
```

```python
# 2. Compute Systemic Tau with Numba Optimization
res = systemic_tau(X, window_size=13, stride=1)
print(f"Computed in {res.metadata['computation_time_seconds']:.3f} seconds")

from systemictau.visualization import plot_tau_evolution
plot_tau_evolution(res.taus_global)
```

```python
# 3. Extract Ontological Ascent Layers
hp_z, core_hyper = hyper_persistence(res.taus_global)
lam, tt = rolling_rqa(res.taus_global)
M = critical_mass_metric(hp_z, lam, tt)
A = compute_antisynchronization(res.taus_per_module)

episodes = extract_joint_episodes(A, M, theta_A=0.04, theta_M=1.0, D_min=30)
print(f"Detected {len(episodes)} Joint Episodes")

from systemictau.visualization import plot_joint_episodes
plot_joint_episodes(res.taus_global, episodes)
```
